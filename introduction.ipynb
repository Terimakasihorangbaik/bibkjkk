```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
datas = pd.read_csv("data/cause_of_death_indonesia.csv")
```

```python
datas.head()
# head function is to show the first 5 rows of the data
```

```python
datas.shape
# shape function is to show the number of rows and columns
```

```python
datas.info()
#info function is to show the information of the data spesiically the data columns
```

```python
datas.describe()
#describe function is to show the statistical information of the data
```

```python
datas["Total Deaths"].describe()
#describe function is to show the statistical information of the data (but above is spesific one column)
```

```python
datas["Total Deaths"].mean()
```

```python
datas["Total Deaths"].plot(kind="box", vert = False, figsize=(14,6))
# plot function is to show the plot of the data
# kind is to show the type of the plot
# vert is to show the vertical or horizontal plot
# figsize is to show the size of the plot
```

```python
datas["Total Deaths"].plot(kind='density', figsize=(14,6))
```

```python
ax = datas["Total Deaths"].plot(kind='density', figsize=(14,6))
ax.axvline(datas["Total Deaths"].mean(), color='red')
ax.axvline(datas["Total Deaths"].median(), color='green')
```

```python
ax = datas["Total Deaths"].plot(kind='hist', figsize=(14,6))
ax.set_ylabel('Number of Deaths')
ax.set_xlabel('Number of Causes')
```

```python
datas["Year"].value_counts()
# value counts function is to show the number of the data in each column
```

```python
datas["Year"].value_counts().plot(kind='pie', figsize=(14,6))
```

```python
datas.plot(kind='scatter', x='Year', y='Total Deaths', figsize=(14,6))
```

```python
datas.loc[datas["Year"] == 2010, "Total Deaths"].mean()
```

```python
datas.loc[datas['Cause'] == 'AIDS', 'Total Deaths']
# function above is to show the data that have the spesific value in the column
```

```python
datas.loc[(datas['Cause'] == 'AIDS') & (datas['Year'] == 2010), 'Total Deaths']
# we can use loc function multiple times
```
